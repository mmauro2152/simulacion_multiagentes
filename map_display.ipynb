```python
from matplotlib import pyplot as plt
from utils.model import car_model
from utils.plot_utils import lane_width, plot_map, plot_semaforos
```

```python
params = {
        "steps": 2000, 
        "size": 50, 
        "lane_width": lane_width,
        "mode": "fixed",
        "follow_sequence": True
    }
    
model = car_model(params)
model.setup()
```

```python

def show_all_routes():
    axx = [0, 0, 1, 1]
    axy = [0, 1, 1, 0]

    fig, axs = plt.subplots(2, 2)
    
    for i in range(4):
        model.semaforos_manager.semaforos[(i - 1) % 4].change_phase("red")
        model.semaforos_manager.semaforos[i].change_phase("green")

        curr_ax = axs[axx[i], axy[i]]
        plot_map(curr_ax, model.size)
        plot_semaforos(curr_ax, model.semaforos_manager.semaforos, False)

        for r in model.routes[i]:
            curr_ax.plot(r.x, r.y, model.route_colors[i], linestyle="--", linewidth=0.4)

show_all_routes()
```
